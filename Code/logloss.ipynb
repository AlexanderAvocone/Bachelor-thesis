# Log loss for different estimators

## 3.0 GeV

```python
import os
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from xgboost import XGBClassifier


#--------------u3_10000--------------------------------------------------

model = XGBClassifier()
model.load_model("/work/aavocone/models/u3_0_1000.txt")


loss_0 = model.evals_result()["validation_0"]["logloss"]
loss_1 = model.evals_result()["validation_1"]["logloss"]
plt.figure(figsize=(9,6))
x = range(len(loss_1))
plt.plot(x,loss_0, label = "train")
plt.plot(x,loss_1, label = "test")
plt.xlabel("number of trees", fontsize = 20)
plt.ylabel("logloss",fontsize=15)
plt.title(f"Undersampled", fontsize = 15)
plt.legend()
plt.show()

minimum1 = [loss_1.index(min(loss_1)),min(loss_1)]

#------------------o3_1000-----------------------------------------------

model.load_model("/work/aavocone/models/o3_0_1000.txt")
loss_0 = model.evals_result()["validation_0"]["logloss"]
loss_1 = model.evals_result()["validation_1"]["logloss"]
plt.figure(figsize=(9,6))
x = range(len(loss_1))
plt.plot(x,loss_0, label = "train")
plt.plot(x,loss_1, label = "test")
plt.xlabel("number of trees", fontsize = 20)
plt.ylabel("logloss",fontsize=15)
plt.title(f"Oversampled", fontsize = 15)
plt.legend()


minimum2 = [loss_1.index(min(loss_1)),min(loss_1)]

#-------------------------------3_141----------------------------------------


model.load_model("/work/aavocone/models/3_0_141.txt")
loss_0 = model.evals_result()["validation_0"]["logloss"]
loss_1 = model.evals_result()["validation_1"]["logloss"]
plt.figure(figsize=(9,6))
x = range(len(loss_1))
plt.plot(x,loss_0, label = "train")
plt.plot(x,loss_1, label = "test")
plt.xlabel("number of trees", fontsize = 20)
plt.ylabel("logloss",fontsize=15)
plt.title(f"Oversampled", fontsize = 15)
plt.legend()


minimum3 = [loss_1.index(min(loss_1)),min(loss_1)]

print(minimum1)
print(minimum2)
print(minimum3)
#-------------------------------3_500-----------------------------------

model.load_model("/work/aavocone/models/3_0_500.txt")
loss_0 = model.evals_result()["validation_0"]["logloss"]
loss_1 = model.evals_result()["validation_1"]["logloss"]
plt.figure(figsize=(9,6))
x = range(len(loss_1))
plt.plot(x,loss_0, label = "train")
plt.plot(x,loss_1, label = "test")
plt.xlabel("number of trees", fontsize = 20)
plt.ylabel("logloss",fontsize=15)
plt.title(f"Oversampled", fontsize = 15)
plt.legend()
```
