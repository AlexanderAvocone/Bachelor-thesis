# Log loss for different estimators

## 0.005 GeV

```python
import os
import numpy as np
import xgboost as xgb
import matplotlib.pyplot as plt
from xgboost import XGBClassifier


model = XGBClassifier()
model.load_model("/ceph/aavocone/models/new_0_0_model500.txt")
print(model.best_ntree_limit)


loss_0 = model.evals_result()["validation_0"]["logloss"]
loss_1 = model.evals_result()["validation_1"]["logloss"]
plt.figure(figsize=(9,6))
x = range(len(loss_1))
plt.plot(x,loss_0, label = "train")
plt.plot(x,loss_1, label = "test")
plt.xlabel("number of trees", fontsize = 20)
plt.ylabel("logloss",fontsize=15)
plt.title(f"original", fontsize = 15)
plt.legend()
plt.savefig(f"/work/aavocone/week11/logloss/ll_0_original.pdf", format="pdf",bbox_inches="tight")
plt.show()

#-----------------------------------------------------------------------------------------------------


model.load_model("/ceph/aavocone/models/under0_500.txt")
print(model.best_ntree_limit)
loss_0 = model.evals_result()["validation_0"]["logloss"]
loss_1 = model.evals_result()["validation_1"]["logloss"]
minimum1 = [loss_1.index(min(loss_1)),min(loss_1)]


plt.figure(figsize=(9,6))
x = range(len(loss_1))
plt.plot(x,loss_0, label = "train")
plt.plot(x,loss_1, label = f"test, minimum = {minimum1[0]}")
plt.xlabel("number of trees", fontsize = 20)
plt.ylabel("logloss",fontsize=15)
plt.title(f"underampled", fontsize = 15)
plt.legend()
plt.savefig(f"/work/aavocone/week11/logloss/loss_0_under.pdf", format="pdf",bbox_inches="tight")
plt.show()

#---------------------------------------------------------------------------------------------------------


model.load_model("/ceph/aavocone/models/over0_500.txt")
print(model.best_ntree_limit)


loss_0 = model.evals_result()["validation_0"]["logloss"]
loss_1 = model.evals_result()["validation_1"]["logloss"]
minimum1 = [loss_1.index(min(loss_1)),min(loss_1)]
plt.figure(figsize=(9,6))
x = range(len(loss_1))
plt.plot(x,loss_0, label = "train")
plt.plot(x,loss_1, label = f"test")
plt.xlabel("number of trees", fontsize = 20)
plt.ylabel("logloss",fontsize=15)
plt.title(f"oversampled", fontsize = 15)
plt.legend()
plt.savefig(f"/work/aavocone/week11/logloss/loss_0_over.pdf", format="pdf",bbox_inches="tight")
plt.show()

print(min(loss_1), loss_1[-1])
```
